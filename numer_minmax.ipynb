```python
import pandas as pd
import numpy as np
import os

os.chdir("/Users/PraveenGupta/Downloads/Kaggle/numerai_datasets/")

train=pd.read_csv("numerai_training_data.csv")
test=pd.read_csv("numerai_tournament_data.csv")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Normalizer, PolynomialFeatures
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler, StandardScaler, RobustScaler
```

```python
feature_names = train.columns.values
train_id_target = train[['t_id','target']]
train = train.drop( 'target', axis = 1 )
train = train.drop( 't_id', axis = 1 )
test_target=test[['t_id']]
test = test.drop( 't_id', axis = 1 )
```

```python
print train.columns.values
```

```python
transformers = [MaxAbsScaler(), MinMaxScaler(), RobustScaler(), StandardScaler(),Normalizer( norm = 'l1' ), Normalizer( norm = 'l2' ), Normalizer( norm = 'max' )] 
transformationNames=['MaxAbsScaler','MinMaxScaler','RobustScaler','StandardScaler','Normalizer_L1','Normalizer_L2','Normalizer_max']
```

```python
def transform_train_and_evaluate(transformer):
    train_new = transformer.fit_transform(train)#,train_id_target['target'].values)        #do we need to pass the target here??
    test_new = transformer.transform(test)
    train_new = pd.DataFrame(train_new,columns=train.columns.values)
    train_new = pd.concat([train_new,train_id_target],axis=1)
    test_new = pd.DataFrame(test_new,columns=test.columns.values)
    test_new = pd.concat([test_new,test_target],axis=1)
    return train_new,test_new
```

```python
i=0
for transformer in transformers:
    train_transformed,test_transformed = transform_train_and_evaluate(transformer)
    train_transformed.to_csv(transformationNames[i]+"_train.csv",index=False)
    test_transformed.to_csv(transformationNames[i]+"_test.csv",index=False)
    print "created new table : %s "%((str(transformer)))
    i=i+1
```
